```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import os
# Для відтворюваності
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)
# Параметри
BATCH_SIZE = 32
EPOCHS = 5
IMG_SHAPE = (28, 28, 1)
NUM_CLASSES = 10
LR = 0.001
# Завантаження MNIST
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = (train_images.astype('float32') / 255.0).reshape(-1, 28, 28, 1)
test_images = (test_images.astype('float32') / 255.0).reshape(-1, 28, 28, 1)
print("Train:", train_images.shape, train_labels.shape)
print("Test:", test_images.shape, test_labels.shape)
```

```python
def build_cnn_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=IMG_SHAPE),
        tf.keras.layers.Conv2D(filters=24, kernel_size=(3,3),
activation='relu'),
        tf.keras.layers.MaxPool2D(pool_size=(2,2)),
        tf.keras.layers.Conv2D(filters=36, kernel_size=(3,3),
activation='relu'),
        tf.keras.layers.MaxPool2D(pool_size=(2,2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    return model
```

```python
model_sparse = build_cnn_model()
opt = tf.keras.optimizers.Adam(learning_rate=LR)
model_sparse.compile(optimizer=opt,
loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history_sparse = model_sparse.fit(
train_images, train_labels,
batch_size=BATCH_SIZE, epochs=EPOCHS,
validation_split=0.1, verbose=2
)
test_loss_sparse, test_acc_sparse = model_sparse.evaluate(test_images, test_labels, verbose=0)
print("Sparse loss → Test accuracy:", test_acc_sparse)
```

```python
# --- MLP (fully connected) ---
def build_mlp_model():
    return tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28, 1)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

mlp = build_mlp_model()
mlp.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LR),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_mlp = mlp.fit(
    train_images, train_labels,
    batch_size=BATCH_SIZE, epochs=EPOCHS,
    validation_split=0.1, verbose=2
)

mlp_test_loss, mlp_test_acc = mlp.evaluate(test_images,
test_labels, verbose=0)
print("MLP → Test accuracy:", mlp_test_acc)
```

```python
import os, matplotlib.pyplot as plt
os.makedirs("figs", exist_ok=True)
def plot_and_save_history(history, title, fname_prefix):
  acc = history.history.get('accuracy', [])
  val_acc = history.history.get('val_accuracy', [])
  loss = history.history.get('loss', [])
  val_loss = history.history.get('val_loss', [])
  epochs_range = range(1, len(acc)+1)

  plt.figure(figsize=(12,5))

  # Accuracy
  plt.subplot(1,2,1)
  plt.plot(epochs_range, acc, label='Train acc')
  if val_acc: plt.plot(epochs_range, val_acc, label='Val acc')
  plt.xlabel('Epoch'); plt.ylabel('Accuracy')
  plt.title(f'Accuracy — {title}')
  plt.legend(loc='lower right')

  # Loss
  plt.subplot(1,2,2)
  plt.plot(epochs_range, loss, label='Train loss')
  if val_loss: plt.plot(epochs_range, val_loss, label='Val loss')
  plt.xlabel('Epoch'); plt.ylabel('Loss')
  plt.title(f'Loss — {title}')
  plt.legend(loc='upper right')

  plt.tight_layout()
  out = f"figs/{fname_prefix}_history.png"
  plt.savefig(out, dpi=150, bbox_inches='tight')
  plt.show()
  print(f"Saved: {out}")

# Виклики для двох історій:
plot_and_save_history(history_mlp,    "MLP (Adam, bs=32, ep=5)", "mlp")
plot_and_save_history(history_sparse, "CNN (Adam, bs=32, ep=5)"  ,  "cnn_sparse")
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf

activations = ['relu', 'leakyrelu', 'tanh', 'sigmoid']
histories = {}

def build_cnn_model(activation='relu'):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=IMG_SHAPE))

    # Conv + Pool 1
    model.add(tf.keras.layers.Conv2D(filters=24, kernel_size=(3,3)))
    if activation.lower() == 'leakyrelu':
        model.add(tf.keras.layers.LeakyReLU(alpha=0.1))
    else:
        model.add(tf.keras.layers.Activation(activation))
    model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

    # Conv + Pool 2
    model.add(tf.keras.layers.Conv2D(filters=36, kernel_size=(3,3)))
    if activation.lower() == 'leakyrelu':
        model.add(tf.keras.layers.LeakyReLU(alpha=0.1))
    else:
        model.add(tf.keras.layers.Activation(activation))
    model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

    # Dense
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(128))
    if activation.lower() == 'leakyrelu':
        model.add(tf.keras.layers.LeakyReLU(alpha=0.1))
    else:
        model.add(tf.keras.layers.Activation(activation))

    model.add(tf.keras.layers.Dense(NUM_CLASSES, activation='softmax'))
    return model

# --- Тренування для всіх активацій ---
for act in activations:
    print(f"\n=== Training CNN with activation: {act} ===")
    model = build_cnn_model(act)
    opt = tf.keras.optimizers.Adam(learning_rate=LR)
    model.compile(optimizer=opt,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(train_images, train_labels,
                        batch_size=BATCH_SIZE, epochs=EPOCHS,
                        validation_split=0.1, verbose=2)

    test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
    print(f"{act} → Test accuracy: {test_acc:.4f}")

    histories[act] = history.history

# --- Графік точності ---
plt.figure(figsize=(10,6))
for act, hist in histories.items():
    plt.plot(hist['accuracy'], linestyle='-', label=f'{act} (train)')
    plt.plot(hist['val_accuracy'], linestyle='--', label=f'{act} (val)')
plt.title('Training and Validation Accuracy for Different Activations')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# --- Графік втрат ---
plt.figure(figsize=(10,6))
for act, hist in histories.items():
    plt.plot(hist['loss'], linestyle='-', label=f'{act} (train)')
    plt.plot(hist['val_loss'], linestyle='--', label=f'{act} (val)')
plt.title('Training and Validation Loss for Different Activations')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf

os.makedirs("figs", exist_ok=True)

activations = ['relu', 'leakyrelu', 'tanh', 'sigmoid']
histories = {}

# --- Модель CNN ---
def build_cnn_model(activation='relu'):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=IMG_SHAPE))

    model.add(tf.keras.layers.Conv2D(filters=24, kernel_size=(3,3)))
    if activation.lower() == 'leakyrelu':
        model.add(tf.keras.layers.LeakyReLU(alpha=0.1))
    else:
        model.add(tf.keras.layers.Activation(activation))
    model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

    model.add(tf.keras.layers.Conv2D(filters=36, kernel_size=(3,3)))
    if activation.lower() == 'leakyrelu':
        model.add(tf.keras.layers.LeakyReLU(alpha=0.1))
    else:
        model.add(tf.keras.layers.Activation(activation))
    model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))

    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(128))
    if activation.lower() == 'leakyrelu':
        model.add(tf.keras.layers.LeakyReLU(alpha=0.1))
    else:
        model.add(tf.keras.layers.Activation(activation))

    model.add(tf.keras.layers.Dense(NUM_CLASSES, activation='softmax'))
    return model

# --- Функція для побудови та збереження графіків (твоя) ---
def plot_and_save_history(history, title, fname_prefix):
    acc = history.history.get('accuracy', [])
    val_acc = history.history.get('val_accuracy', [])
    loss = history.history.get('loss', [])
    val_loss = history.history.get('val_loss', [])
    epochs_range = range(1, len(acc)+1)

    plt.figure(figsize=(12,5))

    # Accuracy
    plt.subplot(1,2,1)
    plt.plot(epochs_range, acc, label='Train acc')
    if val_acc: plt.plot(epochs_range, val_acc, label='Val acc')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy')
    plt.title(f'Accuracy — {title}')
    plt.legend(loc='lower right')

    # Loss
    plt.subplot(1,2,2)
    plt.plot(epochs_range, loss, label='Train loss')
    if val_loss: plt.plot(epochs_range, val_loss, label='Val loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss')
    plt.title(f'Loss — {title}')
    plt.legend(loc='upper right')

    plt.tight_layout()
    out = f"figs/{fname_prefix}_history.png"
    plt.savefig(out, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {out}")

# --- Тренування для всіх активацій ---
for act in activations:
    print(f"\n=== Training CNN with activation: {act} ===")
    model = build_cnn_model(act)
    opt = tf.keras.optimizers.Adam(learning_rate=LR)
    model.compile(optimizer=opt,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(train_images, train_labels,
                        batch_size=BATCH_SIZE, epochs=EPOCHS,
                        validation_split=0.1, verbose=2)

    test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
    print(f"{act} → Test accuracy: {test_acc:.4f}")

    histories[act] = history

    # --- Побудова графіка для цієї активації ---
    plot_and_save_history(history, f"CNN ({act})", f"cnn_{act}")

# --- Порівняльний графік для всіх активацій ---
plt.figure(figsize=(12,5))

# Accuracy
plt.subplot(1,2,1)
for act, hist in histories.items():
    plt.plot(range(1, len(hist.history['accuracy'])+1),
             hist.history['val_accuracy'],
             label=f'{act}')
plt.title('Validation Accuracy — All Activations')
plt.xlabel('Epoch'); plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss
plt.subplot(1,2,2)
for act, hist in histories.items():
    plt.plot(range(1, len(hist.history['loss'])+1),
             hist.history['val_loss'],
             label=f'{act}')
plt.title('Validation Loss — All Activations')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
